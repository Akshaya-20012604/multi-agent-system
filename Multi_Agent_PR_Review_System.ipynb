```python
import os

all_code = ""

for root, dirs, files in os.walk("project"):

    for file in files:

        if file.endswith(".java"):

            file_path = os.path.join(root, file)

            try:
                with open(file_path, "r", encoding="utf-8") as f:

                    code = f.read()

                    all_code += f"\n\n===== FILE: {file_path} =====\n"
                    all_code += code

            except Exception as e:
                print(f"Error reading {file_path}: {e}")

print("Java project loaded successfully")
```

```python
def generate_response(prompt):

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )

    return response.text
```

```python
def run_security_agent(code):

    security_prompt = f"""
    You are a senior application security engineer.

    Analyze the following Java application.

    Detect:
    - SQL Injection
    - Hardcoded secrets
    - Authentication flaws
    - Sensitive logging
    - Insecure APIs
    - Unsafe deserialization
    - Broken access control
    - Security misconfigurations

    Return ONLY structured JSON.

    Format:

    {{
      "summary": {{
        "critical": 0,
        "high": 0,
        "medium": 0,
        "low": 0
      }},
      "findings": [
        {{
          "issue": "",
          "severity": "",
          "file": "",
          "line": "",
          "description": "",
          "fix": ""
        }}
      ]
    }}

    Codebase:
    {code}
    """

    return generate_response(security_prompt)
```

```python
def run_quality_agent(code):

    quality_prompt = f"""
    You are a senior Java code reviewer.

    Analyze the Java application.

    Detect:
    - bad naming conventions
    - dead code
    - unused variables
    - Java best practice violations
    - maintainability issues
    - incorrect string comparisons
    - poor readability
    - bad architecture patterns
    - duplicate code

    Return ONLY structured JSON.

    Format:

    {{
      "summary": {{
        "high": 0,
        "medium": 0,
        "low": 0
      }},
      "findings": [
        {{
          "issue": "",
          "severity": "",
          "file": "",
          "line": "",
          "description": "",
          "fix": ""
        }}
      ]
    }}

    Codebase:
    {code}
    """

    return generate_response(quality_prompt)
```

```python
def run_test_agent(code):

    test_prompt = f"""
    You are a senior QA engineer.

    Analyze the Java application.

    Generate:
    - missing test cases
    - edge cases
    - null input tests
    - authentication tests
    - SQL injection tests
    - performance tests
    - API failure tests

    Return ONLY structured JSON.

    Format:

    {{
      "summary": {{
        "total_tests_suggested": 0
      }},
      "test_cases": [
        {{
          "scenario": "",
          "severity": "",
          "file": "",
          "description": "",
          "expected_result": ""
        }}
      ]
    }}

    Codebase:
    {code}
    """

    return generate_response(test_prompt)
```

```python
def run_aggregator(
    security_output,
    quality_output,
    test_output
):

    aggregation_prompt = f"""
    You are a senior engineering manager.

    Combine all findings into a professional engineering review report.

    Generate:

    1. Executive Summary
    2. Security Findings
    3. Code Quality Findings
    4. Test Recommendations
    5. Severity Breakdown
    6. Final Engineering Recommendation

    Return in professional markdown format.

    Security Findings:
    {security_output}

    Quality Findings:
    {quality_output}

    Test Findings:
    {test_output}
    """

    return generate_response(aggregation_prompt)
```

```python
print("Running Security Agent...")
security_output = run_security_agent(all_code)

print("Running Quality Agent...")
quality_output = run_quality_agent(all_code)

print("Running Test Agent...")
test_output = run_test_agent(all_code)

print("Running Aggregator...")
final_review = run_aggregator(
    security_output,
    quality_output,
    test_output
)

print("Review Completed Successfully")
```
